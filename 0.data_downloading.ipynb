```python
import subprocess
import os
from glob import glob
```

```python
# Define the dataset directory
data_dir = './ds003523'
# Navigate to the dataset directory
os.chdir(data_dir)
```

```python
# Find all subject directories

subject_dirs = glob("sub-*")

# Download data for each subject
for subject_dir in subject_dirs:
    print(f"Downloading data for {subject_dir}...")
    try:
        subprocess.run(["datalad", "get", subject_dir], check=True)
        print(f"Data for {subject_dir} has been downloaded.")
    except subprocess.CalledProcessError as e:
        print(f"Error downloading data for {subject_dir}: {e}")
```

```python
import re
subject_dirs.sort(key=lambda x: int(re.search(r'\d+', x).group()))
 
# Limit to the top 10 subject directories
top_subject_dirs = subject_dirs[:10]
 
# Download data for each subject in the top 10
for subject_dir in top_subject_dirs:
    print(f"Downloading data for {subject_dir}...")
    try:
        subprocess.run(["datalad", "get", subject_dir], check=True)
        print(f"Data for {subject_dir} has been downloaded.")
    except subprocess.CalledProcessError as e:
        print(f"Error downloading data for {subject_dir}: {e}")
```
